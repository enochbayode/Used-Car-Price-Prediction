```python
def accident_col(df):
    Data['accident_dummy'] = Data['accident'].apply(lambda x: 0 if x == 'NONE REPORTED' else (1 if x == 'AT LEAST 1 ACCIDENT OR DAMAGE REPORTED' else None))
    return Data
```

```python
def transmission_col(df):
    def classify_transmission(transmission):
        if 'M/T' in transmission or 'MT' in transmission or 'MANUAL' in transmission:
            return 'M/T'
        elif 'A/T' in transmission or 'AT' in transmission or 'AUTOMATIC' in transmission or 'CVT TRANSMISSION' in transmission:
            return 'A/T'
        else:
            return 'OTHER'

    Data['transmission'] = Data['transmission'].apply(classify_transmission)
    return df
```

```python
def ext_and_int_col(df):

  Data['ext_col'] = Data['ext_col'].apply(lambda x: x if x in ["BLACK","WHITE","GRAY","SILVER","BLUE","RED","ORANGE","SNOWFLAKE WHITE PEARL METALLIC",
                                                               "RADIANT RED METALLIC II","GREEN","OBSIDIAN BLACK METALLIC" ] else 'OTHERS')

  Data['int_col'] = Data['int_col'].apply(lambda x: x if x in ['BLACK', "BEIGE", "GRAY","WHITE","BLUE","BROWN","GRAPHITE"] else 'OTHERS')

  return Data
```

```python
def fuel_type_col(Data):

    # Data['fuel_type'] = Data['fuel_type'].replace({'HYBRID': 'HYBRID', 'PLUG-IN HYBRID': 'HYBRID','NOT SUPPORTED':'OTHER','–':'OTHER'})
    # return Data

    Data['fuel_type'] = Data['fuel_type'].replace({
        'GASOLINE/MILD ELECTRIC HYBRID': 'HYBRID',
        'GAS/ELECTRIC HYBRID': 'HYBRID',
        'PLUG-IN HYBRID': 'HYBRID',
        'HYBRID': 'HYBRID',
        'NOT SUPPORTED': 'OTHER',
        '–': 'OTHER'
    })

# I kept "Gasoline", "Diesel", "Electric", "Flex Fuel Capability" unchanged since they’re already clear.
```

```python
def engine_col(df):
    # Create hp column
    Data['hp'] = Data['engine'].str.extract(r'(\d+\.\d+)HP').astype(float, errors='ignore')

    # Create engine displacement
    Data['engine displacement'] = Data['engine'].str.extract(r'(\d+\.\d+)L')
    Data['engine displacement'] = Data['engine displacement'].fillna(Data['engine'].str.extract(r'(\d+\.\d+)LITER')[0])
    Data['engine displacement'] = Data['engine displacement'].astype(float, errors='ignore')

    # Create cylinder
    Data['cylinder'] = Data['engine'].str.extract(r'(\d+) CYLINDER')
    Data['cylinder'] = Data['cylinder'].fillna(Data['engine'].str.extract(r'V(\d)')[0])
    Data['cylinder'] = Data['cylinder'].astype('Int64', errors='ignore')

    # Create fuel type
    Data['fuel'] = Data['engine'].str.extract(r'(GASOLINE|DIESEL|ELECTRIC|HYBRID|GASOLINE/MILD ELECTRIC HYBRID|GAS/ELECTRIC HYBRID|FLEX FUEL CAPABILITY)')
    Data['fuel_type'] = Data['fuel_type'].combine_first(Data['fuel'])

    # Is it V type or not
    Data['is_v_engine'] = Data['engine'].str.contains(r'V\d+', case=False, na=False)

    # Create turbo
    Data['turbo'] = Data['engine'].str.contains('TWIN TURBO', case=False, na=False)

    # Create dohc
    Data['dohc'] = Data['engine'].str.contains('DOHC', case=False, na=False)

    # Create MPFI
    Data['mpfi'] = Data['engine'].str.contains('MPFI', case=False, na=False)

    # Create GDI
    Data['gdi'] = Data['engine'].str.contains('GDI', case=False, na=False)

    # Create PDI
    Data['pdi'] = Data['engine'].str.contains('PDI', case=False, na=False)

    # Create OHC
    Data['ohc'] = Data['engine'].str.contains('OHC', case=False, na=False)


    return Data
```

```python
def fill_na(df, bugatti_avg_hp_external=1300):
    Data['hp'] = Data.groupby('brand')['hp'].transform(lambda x: x.fillna(x.mean()))
    Data.loc[(df['brand'] == 'BUGATTI') & (Data['hp'].isnull()), 'hp'] = bugatti_avg_hp_external

    most_common_fuel = df.groupby('brand')['fuel_type'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
    most_common_displacement = df.groupby('brand')['engine displacement'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
    most_common_cylinder = df.groupby('brand')['cylinder'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

    Data['fuel_type'] = Data.apply(
        lambda row: most_common_fuel[row['brand']] if pd.isna(row['fuel_type']) else row['fuel_type'],
        axis=1
    )

    Data['engine displacement'] = Data.apply(
        lambda row: most_common_displacement[row['brand']] if pd.isna(row['engine displacement']) else row['engine displacement'],
        axis=1
    )

    Data['cylinder'] = Data.apply(
        lambda row: most_common_cylinder[row['brand']] if pd.isna(row['cylinder']) else row['cylinder'],
        axis=1
    )

    Data['cylinder'] = Data.apply(
        lambda row: most_common_cylinder[row['brand']] if pd.isna(row['cylinder']) else row['cylinder'],
        axis=1
    )
    Data['accident_dummy'].fillna(0, inplace=True)

    if Data['cylinder'].isnull().any():
        most_common_cylinder_value = Data['cylinder'].mode()[0]
        Data['cylinder'].fillna(most_common_cylinder_value, inplace=True)

    return Data
```

```python
def one_hot_encoding(Data):
    Data = pd.get_dummies(Data, columns=['brand', 'fuel_type', 'transmission', 'ext_col', 'int_col'], drop_first=False)
    return Data
```

```python
def age_and_usage(Data):
    current_year = 2025
    Data['age'] = current_year - Data['model_year']

    Data['annual_km'] = Data.apply(
        lambda row: row['milage'] / row['age'] if row['age'] > 0 else row['milage'],
        axis=1
    )

    return Data
```

```python
def clean_strings(Data):
    for col in Data.select_dtypes(include="object"):
        Data[col] = Data[col].astype(str).str.upper()
    return Data
```

```python
def all_processing(Data, bugatti_avg_hp_external=800):
    Data = Data.drop(columns=['id', 'model', 'clean_title'])

    Data = Data.applymap(lambda x: x.upper() if isinstance(x, str) else x)

    # Apply uppercase only to object type columns
    # for col in Data.select_dtypes(include='object').columns:
    #     Data[col] = Data[col].apply(lambda x: x.upper() if isinstance(x, str) else x)

    Data = clean_strings(Data)

    Data = accident_col(Data)

    Data = transmission_col(Data)

    Data = ext_and_int_col(Data)

    Data = fuel_type_col(Data)

    Data = engine_col(Data)

    Data = fill_na(Data, bugatti_avg_hp_external)

    Data = one_hot_encoding(Data)

    Data = age_and_usage(Data)


    return Data
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
def model_create(Data, target='price', model_type='lightgbm'):
    X = Data.drop([target], axis=1)
    y = Data[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    if  model_type == 'xgboost':
        model = XGBRegressor()
    elif model_type == 'lightgbm':
        model = LGBMRegressor(verbosity=-1)
    elif model_type == 'catboost':
        model = CatBoostRegressor(verbose=0)
    else:
        raise ValueError("Change it. ")
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    return model, X_train, X_test, y_train, y_test, y_pred
```

```python
from sklearn.metrics import mean_squared_error

def calculate_rmse(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    return rmse
```

```python
from bayes_opt import BayesianOptimization
from skopt import BayesSearchCV

def bayesian_hyperparameter_optimization(X_train, y_train, model_type):
    if model_type == 'lightgbm':
        model = LGBMRegressor(verbosity=-1)
        search_spaces = {
            'n_estimators': (30, 300),
            'learning_rate': (0.01, 0.2, 'log-uniform'),
            'num_leaves': (10, 100),
            'max_depth': (-1, 15),
        }

    elif model_type == 'xgboost':
        model = XGBRegressor()
        search_spaces = {
            'n_estimators': (100, 500),
            'learning_rate': (0.01, 0.3, 'log-uniform'),
            'max_depth': (3, 15),
            'min_child_weight': (1, 10)
        }

    elif model_type == 'catboost':
        model = CatBoostRegressor(verbose=0)
        search_spaces = {
            'iterations': (100, 500),
            'learning_rate': (0.01, 0.3, 'log-uniform'),
            'depth': (3, 15),
            'l2_leaf_reg': (1, 10)
        }

    else:
        raise ValueError("Change it.")

    opt = BayesSearchCV(model, search_spaces, n_iter=32, cv=5, scoring='neg_mean_squared_error', n_jobs=-1,random_state=42)
    opt.fit(X_train, y_train)

    return opt.best_estimator_, opt.best_params_
```

```python
best_model, best_params = bayesian_hyperparameter_optimization(X_train, y_train, model_type='lightgbm')
print("The best model:", best_model)
print("The best parameters:", best_params)
```
